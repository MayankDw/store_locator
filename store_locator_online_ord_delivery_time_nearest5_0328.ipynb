```python
from sklearn import cluster
import numpy as np
import folium
from geopy.distance import great_circle

#store_points = np.array(list(store_geodata.values()))
store_points = store_coord

# Assign cluster size and apply kmeans
cluster_size = 52
k_means = cluster.KMeans(n_clusters=cluster_size, random_state = 12)
k_means.fit(store_points)
centers = (k_means.cluster_centers_)
label = k_means.labels_

# Create Folium map centered around the midpoint of points
map_center = [np.mean(store_points[:, 0]), np.mean(store_points[:, 1])]
mapwa = folium.Map(location=map_center, zoom_start=10)

# Assign points to clusters and calculate circle radius
for cluster in range(cluster_size):
    cluster_points = store_points[label == cluster]
    center = centers[cluster]
    folium.Marker(location=center.tolist(), popup=f'Cluster {cluster}', icon=folium.Icon(color='red', icon="star")).add_to(mapwa)

    # Calculate the maximum distance to a point in the cluster
    max_distance = max(great_circle(center, point).meters for point in cluster_points)

    # Draw circle around cluster center covering all cluster points
    folium.Circle(location=center.tolist(), radius=max_distance, color='blue', fill=True, fill_opacity=0.2).add_to(mapwa)

    # Add cluster points to the map
    for point in cluster_points:
        folium.Marker(location=point.tolist(), icon=folium.Icon(color="green"), fill=True).add_to(mapwa)

# Save map to HTML file
mapwa.save("kmeans_clusters_mapw.html")
mapwa
```

```python
import geopy.distance
from geopy.geocoders import Nominatim
from folium.plugins import AntPath


def find_nearest_locations(address, locations, num_nearest=5):
    geolocator = Nominatim(user_agent="geoapi")
    location = geolocator.geocode(address)
    if not location:
        return None, None

    given_point = (location.latitude, location.longitude)
    distances = [(geopy.distance.geodesic(given_point, loc).mi, loc) for loc in locations]
    distances.sort()
    nearest_locations = [loc for _, loc in distances[:num_nearest]]
    return given_point, nearest_locations

# Define address and find nearest locations
given_address = "25245 Meadowbrook Rd, Novi, MI 48375"
given_point, nearest_points = find_nearest_locations(given_address, store_points)
print(nearest_points)
if given_point:
    # Create Folium map
    # mymap = folium.Map(location=given_point, zoom_start=5)

    # Add given location marker
    # folium.Marker(location=given_point, popup="Given Location", icon=folium.Icon(color='red')).add_to(mymap)

    # Add nearest location markers and draw connecting arcs
    for point in nearest_points:
        #folium.Marker(location=point, popup=f"Nearest Location", icon=folium.Icon(color='blue')).add_to(mymap)
        AntPath([given_point, point], color="blue", weight=2.5, opacity=0.8).add_to(mapwa)

    # Save map
    mapwa.save("nearest_locations_map.html")
    print("Map saved as nearest_locations_map.html")
else:
    print("Address not found!")

mapwa
```
